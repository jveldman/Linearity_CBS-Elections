```python
import os 
import pandas as pd
import sqlite3

# Create list of folders
folders = []
for folder in os.walk('Data/'):
    files = []
    for file in folder: 
        files.append(file)
        #data = pd.read_csv(file)
        #con = sqlite3.connect('D:/data/PolProj.db')
        #data.to_sql(file, con  = con)
    folders.append(files)
        
```

```python

#from sqlite3 import ParserError
con = sqlite3.connect('D:/data/PolProj.db')
for idx, item in enumerate(folders):
    try:
        for i in enumerate(item[2]):
            location = str(str(item[0])+'/'+str(i[1]))
            temp = pd.read_csv(location, sep = ';')
            try:
                temp.to_sql(str(i[1]), con = con, if_exists = 'replace')
            except sqlite3.OperationalError: 
                print('Double colname, check '+str(i[1]))
                next
    except pd.errors.ParserError: 
        print('Something went wrong with '+str(i[1]))
                
            


```
